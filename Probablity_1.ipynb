```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('student-mat.csv')
df.head(3)
```

```python
df.shape
```

##  add a boolean column called grade_a nothing if a student achieved 80% or higher as final score

```python
df['grade_A']=np.where(df['G3']*5>=80,1,0)
```

```python
df.shape
```

## make another boolean column called high_absences with a value of 1 if student missed 10 classes

```python
df['high_absenses']=np.where(df['absences']>=10,1,0)
```

## add one more column to make building a pivot table easier

```python
df['count']=1
```

```python
df
```

```python
df=df[['grade_A','high_absenses','count']]
df.head()
```

```python
pd.pivot_table(df,values='count',index=['grade_A'],columns=['high_absenses'],aggfunc=np.size,fill_value=0)
```

```python
## in our case :P(A) is the prob of grade of 80% or greater. P(b) is the prob of missing 10 or more classes , a:80% + grade 
## Calculations P(A)=(35+5)/(35+5+277+78)=0.101 ,P(B)=(78+5)/(35+5+277+78)=0.2101 , P(A&B)=5/(35+5+277+78) =0.0126 ,,,,
## P(A/B)=0.0126/.2101
###There we have it . The probablity of getting atleast an 80% final grade ,given missing 10 or more classes to 6%
```
